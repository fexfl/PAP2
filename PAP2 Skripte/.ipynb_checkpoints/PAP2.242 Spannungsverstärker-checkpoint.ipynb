```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

#%matplotlib qt5
```

### 1a) Gleichspannungsverstärker

```python
# Messwerte:
# R=47,8 kOhm
U_E_1_478 = np.array([-247, -200, -122, -26.3, 25.4, 119, 199, 248]) /1000 #V
U_E_1_478_err = np.array([1, 1, 1, 0.3, 0.5, 1, 1, 1]) /1000 #V

U_A_1_478 = np.array([4.01, 3.28, 2, 0.487, -0.378, -1.840, -3.120, -3.950]) #V
U_A_1_478_err = np.array([0.01, 0.01, 0.01, 0.003, 0.003, 0.01, 0.01, 0.01]) #V

# R=274kOhm
U_E_1_274 = np.array([-247, -200, -125, -25, 24.6, 119, 200, 247]) /1000 #V
U_E_1_274_err = np.array([1, 1, 1, 0.3, 0.3, 1, 1, 1]) /1000 #V

U_A_1_274 = np.array([14.4, 14.4, 11.4, 2.6, -2, -10.5, -13.6, -13.6]) #V
U_A_1_274_err = np.array([0.1, 0.1, 0.1, 0.01, 0.01, 0.1, 0.1, 0.1]) #V
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.errorbar(U_E_1_478,U_A_1_478, yerr = U_A_1_478_err, xerr = U_E_1_478_err,  linestyle='None', marker = ".", label="Messwerte 47,8 kOhm", capsize = 2, elinewidth=1, color="blue")
plt.errorbar(U_E_1_274,U_A_1_274, yerr = U_A_1_274_err, xerr = U_E_1_274_err,  linestyle='None', marker = ".", label="Messwerte 274 kOhm", capsize = 2, elinewidth=1, color="orange")
plt.xlabel('Eingangsspannung [V]')
plt.ylabel('Ausgangsspannung [V]')
plt.title('Ausgangsspannung als Funktion der Eingangsspannung - Gleichstrom')
#plt.yscale('log')
plt.legend()
plt.grid(linestyle=":", linewidth=1)
```

```python
def linearfun(x,s):
    return x*s #+ b

popt_1_478, pcov_1_478 = curve_fit(linearfun, U_E_1_478, U_A_1_478, sigma=U_A_1_478_err)
popt_1_274, pcov_1_274 = curve_fit(linearfun, U_E_1_274[2:-2], U_A_1_274[2:-2], sigma=U_A_1_274_err[2:-2])
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.errorbar(U_E_1_478,U_A_1_478, yerr = U_A_1_478_err, xerr = U_E_1_478_err,  linestyle='None', marker = ".", label="Messwerte 47,8 kOhm", capsize = 2, elinewidth=1, color="blue")
plt.errorbar(U_E_1_274,U_A_1_274, yerr = U_A_1_274_err, xerr = U_E_1_274_err,  linestyle='None', marker = ".", label="Messwerte 274 kOhm", capsize = 2, elinewidth=1, color="orange")
plt.plot(U_E_1_478,linearfun(U_E_1_478,*popt_1_478), label="fit 47,8 kOhm", color="blue")
plt.plot(U_E_1_274,linearfun(U_E_1_274,*popt_1_274), label="fit 274 kOhm", color="orange")
plt.xlabel('Eingangsspannung [V]')
plt.ylabel('Ausgangsspannung [V]')
plt.title('Ausgangsspannung als Funktion der Eingangsspannung - Gleichstrom')
plt.text(-0.2, -7, 'Fitparameter 47,8 kOhm:')
plt.text(-0.2, -8.5, 's = ' + str(np.round(popt_1_478[0],2)) + '$\pm$' + str(np.round(np.sqrt(pcov_1_478[0,0]),2)))
#plt.text(-0.2, -10, 'b = ' + str(np.round(popt_1_478[1],5)) + '$\pm$' + str(np.round(pcov_1_478[1,1],5)))
plt.text(-0.2, -11.5, 'Fitparameter 274 kOhm:')
plt.text(-0.2, -13, 's = ' + str(np.round(popt_1_274[0],0)) + '$\pm$' + str(np.round(np.sqrt(pcov_1_274[0,0]),0)))
#plt.text(-0.2, -14.5, 'b = ' + str(np.round(popt_1_274[1],4)) + '$\pm$' + str(np.round(pcov_1_274[1,1],4)))
#plt.yscale('log')
plt.legend()
plt.grid(linestyle=":", linewidth=1)
```

```python
# Theoretische Verstärkungen:
R_E_1 = 3000 #kOhm
R_G_1 = np.array([47.8, 274]) *1000 #Ohm

V_0_1 = -R_G_1 / R_E_1
print("Theoretische Verstärkungen:", V_0_1)

# Abweichungen:
V_0_1_exp = np.array([popt_1_478[0], popt_1_274[0]])
V_0_1_exp_err = np.array([np.sqrt(pcov_1_478[0,0]), np.sqrt(pcov_1_274[0,0])])
print(V_0_1_exp_err)
sigma_1 = np.abs((V_0_1_exp - V_0_1)/V_0_1_exp_err)
print("Abweichungen:", sigma_1)
```

### 1b) Wechselspannungsverstärker

```python
# Messwerte:
# R=680 kOhm
U_E_1b_680 = np.array([168, 336, 420, 500, 672, 840]) /1000 /10 #V
U_E_1b_680_err = np.array([2, 2, 1, 1, 2, 1]) /1000 /10  #V

U_A_1b_680 = np.array([3.52, 7.00, 8.72, 10.3, 14.00, 17.4])  #V
U_A_1b_680_err = np.array([0.01, 0.04, 0.06, 0.1, 0.1, 0.1])  #V

# R=274kOhm
U_E_1b_274 = np.array([176, 336, 424, 504, 672, 840]) /1000 /10  #V
U_E_1b_274_err = np.array([1, 2, 3, 1, 1, 1]) /1000 /10  #V

U_A_1b_274 =  np.array([1.46, 2.86, 3.56, 4.26, 5.76, 7.12]) #V
U_A_1b_274_err = np.array([0.01, 0.01, 0.03, 0.01, 0.01, 0.03])  #V
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.errorbar(U_E_1b_680,U_A_1b_680, yerr = U_A_1b_680_err, xerr = U_E_1b_680_err,  linestyle='None', marker = ".", label="Messwerte 680 kOhm", capsize = 2, elinewidth=1, color="red")
plt.errorbar(U_E_1b_274,U_A_1b_274, yerr = U_A_1b_274_err, xerr = U_E_1b_274_err,  linestyle='None', marker = ".", label="Messwerte 274 kOhm", capsize = 2, elinewidth=1, color="green")
plt.xlabel('Eingangsspannung [V]')
plt.ylabel('Ausgangsspannung [V]')
plt.title('Ausgangsspannung als Funktion der Eingangsspannung - Wechselstrom')
#plt.yscale('log')
plt.legend()
plt.grid(linestyle=":", linewidth=1)
```

```python
popt_1b_680, pcov_1b_680 = curve_fit(linearfun, U_E_1b_680, U_A_1b_680, sigma=U_A_1b_680_err)
popt_1b_274, pcov_1b_274 = curve_fit(linearfun, U_E_1b_274, U_A_1b_274, sigma=U_A_1b_274_err)
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.errorbar(U_E_1b_680,U_A_1b_680, yerr = U_A_1b_680_err, xerr = U_E_1b_680_err,  linestyle='None', marker = ".", label="Messwerte 680 kOhm", capsize = 2, elinewidth=1, color="red")
plt.errorbar(U_E_1b_274,U_A_1b_274, yerr = U_A_1b_274_err, xerr = U_E_1b_274_err,  linestyle='None', marker = ".", label="Messwerte 274 kOhm", capsize = 2, elinewidth=1, color="green")
plt.plot(U_E_1b_680,linearfun(U_E_1b_680,*popt_1b_680), label="fit 680 kOhm", color="red")
plt.plot(U_E_1b_274,linearfun(U_E_1b_274,*popt_1b_274), label="fit 274 kOhm", color="green")
plt.xlabel('Eingangsspannung [V]')
plt.ylabel('Ausgangsspannung [V]')
plt.title('Ausgangsspannung als Funktion der Eingangsspannung - Wechselstrom')
plt.text(0.02, 14, 'Fitparameter 680 kOhm:')
plt.text(0.02, 13, 's = ' + str(np.round(popt_1b_680[0],1)) + '$\pm$' + str(np.round(np.sqrt(pcov_1b_680[0,0]),1)))
#plt.text(0.02, 12, 'b = ' + str(np.round(popt_1b_680[1],5)) + '$\pm$' + str(np.round(pcov_1b_680[1,1],5)))
plt.text(0.02, 11, 'Fitparameter 274 kOhm:')
plt.text(0.02, 10, 's = ' + str(np.round(popt_1b_274[0],1)) + '$\pm$' + str(np.round(np.sqrt(pcov_1b_274[0,0]),1)))
#plt.text(0.02, 9, 'b = ' + str(np.round(popt_1b_274[1],4)) + '$\pm$' + str(np.round(pcov_1b_274[1,1],4)))
#plt.yscale('log')
plt.legend()
plt.grid(linestyle=":", linewidth=1)
```

```python
# Theoretische Verstärkungen:
R_E_1b = 3000 #kOhm
R_G_1b = np.array([680, 274]) *1000 #Ohm

V_0_1b = R_G_1b / R_E_1b
print("Theoretische Verstärkungen:", V_0_1b)

# Abweichungen:
V_0_1b_exp = np.array([popt_1b_680[0], popt_1b_274[0]])
V_0_1b_exp_err = np.array([np.sqrt(pcov_1b_680[0,0]), np.sqrt(pcov_1b_274[0,0])])
sigma_1b = np.abs((V_0_1b_exp - V_0_1b)/V_0_1b_exp_err)
print("Abweichungen:", sigma_1b)

# Test
corrfak = V_0_1b[0]/V_0_1b_exp[0]
print("Korrekturfaktor:", corrfak)

V_0_1b_exp_korr = corrfak * V_0_1b_exp
V_0_1b_exp_err_korr = corrfak * V_0_1b_exp_err

print("Korrigiert:", V_0_1b_exp_korr, "+-", V_0_1b_exp_err_korr)
sigma_1b_korr = np.abs((V_0_1b_exp_korr - V_0_1b)/V_0_1b_exp_err_korr)
print("Abweichungen korrigiert:", sigma_1b_korr)
```

## 2) Frequenzgang

```python
# Messwerte:

f_2a = np.array([0.1, 0.4, 0.7, 1, 4, 7, 10, 40, 70, 100, 300]) *1000 #Hz
# R=680 kOhm
U_A_2a_680 = np.array([5.4, 5.4, 5.32, 5.24, 3.92, 2.8, 2.12, 0.68, 0.44, 0.226, 0.08])
U_A_2a_680_err = np.array([0.1, 0.1, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01,0.01, 0.002])

# R=274 kOhm
U_A_2a_274 = np.array([2.16, 2.16, 2.16, 2.16, 2.12, 1.8, 1.56, 0.532, 0.320, 0.222, 0.078])
U_A_2a_274_err = np.array([0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.005])

# R=48,7 kOhm
U_A_2a_487 = np.array([1.3, 1.3, 1.3, 1.26, 1.28, 1.26, 1.26, 1.02, 0.8, 0.64, 0.246])
U_A_2a_487_err = np.array([0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01])

# Tiefpass 
U_A_2b_TP = np.array([1.3, 1.3, 1.3, 1.28, 1.16, 0.88, 0.72, 0.214, 0.127, 0.0906, 0.0324])
U_A_2b_TP_err = np.array([0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.0003, 0.0005])

# Hochpass
U_A_2c_HP = np.array([0.4, 0.766, 0.920, 1.26, 1.27, 1.26, 1.2])
U_A_2c_HP_err = np.array([0.003, 0.010, 0.01, 0.01, .01, 0.01, 0.01])
f_2c = np.array([0.3, 0.7, 1, 4, 7, 10, 20]) *1000 #Hz
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.errorbar(f_2a,U_A_2a_680, yerr = U_A_2a_680_err, linestyle='dashed', marker = ".", label="Widerstand 680 kOhm", capsize = 2, elinewidth=1, color="green")
plt.errorbar(f_2a,U_A_2a_274, yerr = U_A_2a_274_err, linestyle='dashed', marker = ".", label="Widerstand 274 kOhm", capsize = 2, elinewidth=1, color="yellow")
plt.errorbar(f_2a,U_A_2a_487, yerr = U_A_2a_487_err, linestyle='dashed', marker = ".", label="Widerstand 48,7 kOhm", capsize = 2, elinewidth=1, color="crimson")
plt.errorbar(f_2a,U_A_2b_TP, yerr = U_A_2b_TP_err, linestyle='dashed', marker = ".", label="Tiefpass", capsize = 2, elinewidth=1, color="salmon")
plt.errorbar(f_2c,U_A_2c_HP, yerr = U_A_2c_HP_err, linestyle='dashed', marker = ".", label="Hochpass", capsize = 2, elinewidth=1, color="indigo")
plt.xlabel('Frequenz [Hz]')
plt.ylabel('Ausgangsspannung [V]')
plt.title('Ausgangsspannung als Funktion der Frequenz')
plt.yscale('log')
plt.xscale("log")
plt.legend()
plt.grid(linestyle=":", linewidth=1)
```
